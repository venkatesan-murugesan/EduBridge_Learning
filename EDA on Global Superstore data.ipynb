```python
#now we can do our analysis year wise, as we did initially find profit for each category
#first take 2012 data set
pd.pivot_table(data_12,values=['Profit'],index=['Category','Sub-Category'],aggfunc='sum')
```

```python
#now take 2013 data set
pd.pivot_table(data_13,values=['Profit'],index=['Category','Sub-Category'],aggfunc='sum')
```

```python
#now take 2014 data set
pd.pivot_table(data_14,values=['Profit'],index=['Category','Sub-Category'],aggfunc='sum')
```

```python
#now take 2015 data set
pd.pivot_table(data_15,values=['Profit'],index=['Category','Sub-Category'],aggfunc='sum')
```

```python
#now combine each year profit in single pivot table
p12=pd.pivot_table(data_12,values='Profit',index=['Category','Sub-Category'],aggfunc='sum')
p13=pd.pivot_table(data_13,values='Profit',index=['Category','Sub-Category'],aggfunc='sum')
p14=pd.pivot_table(data_14,values='Profit',index=['Category','Sub-Category'],aggfunc='sum')
p15=pd.pivot_table(data_15,values='Profit',index=['Category','Sub-Category'],aggfunc='sum')
```

```python
p1213=pd.merge(p12,p13,on=['Category','Sub-Category'],how='outer',suffixes=('_2012','_2013'))
p1415=pd.merge(p14,p15,on=['Category','Sub-Category'],how='outer',suffixes=('_2014','_2015'))
```

```python
pd.merge(p1213,p1415,on=['Category','Sub-Category'],how='outer')
```

```python
#now take sales pivot
s12=pd.pivot_table(data_12,values='Sales',index=['Category','Sub-Category'],aggfunc='sum')
s13=pd.pivot_table(data_13,values='Sales',index=['Category','Sub-Category'],aggfunc='sum')
s14=pd.pivot_table(data_14,values='Sales',index=['Category','Sub-Category'],aggfunc='sum')
s15=pd.pivot_table(data_15,values='Sales',index=['Category','Sub-Category'],aggfunc='sum')
s1213=pd.merge(s12,s13,on=['Category','Sub-Category'],how='outer',suffixes=('_2012','_2013'))
s1415=pd.merge(s14,s15,on=['Category','Sub-Category'],how='outer',suffixes=('_2014','_2015'))
pd.merge(s1213,s1415,on=['Category','Sub-Category'],how='outer')
```

```python
#now look quantity sold pivot
q12=pd.pivot_table(data_12,values='Quantity',index=['Category','Sub-Category'],aggfunc='sum')
q13=pd.pivot_table(data_13,values='Quantity',index=['Category','Sub-Category'],aggfunc='sum')
q14=pd.pivot_table(data_14,values='Quantity',index=['Category','Sub-Category'],aggfunc='sum')
q15=pd.pivot_table(data_15,values='Quantity',index=['Category','Sub-Category'],aggfunc='sum')
q1213=pd.merge(q12,q13,on=['Category','Sub-Category'],how='outer',suffixes=('_2012','_2013'))
q1415=pd.merge(q14,q15,on=['Category','Sub-Category'],how='outer',suffixes=('_2014','_2015'))
pd.merge(q1213,q1415,on=['Category','Sub-Category'],how='outer')
```
